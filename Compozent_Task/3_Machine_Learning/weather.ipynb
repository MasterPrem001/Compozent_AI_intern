```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('weather_classification_data.csv')
df.head(5)
```

```python
df.columns
```

```python
column = ["Cloud Cover", "Season", "Location", "Weather Type"]
encoder = {}
for col in column:
    encoder[col]= LabelEncoder()
    df[col]= encoder[col].fit_transform(df[col])

df.head(5)
```

```python
x= df.drop(columns=["Temperature", "Weather Type"]) 
y_temp = df['Temperature']
y_weather = df["Weather Type"]
```

```python
x_train, x_test, y_train_temp, y_test_temp = train_test_split(x, y_temp, test_size=0.2, random_state=42)
_, _, y_train_weather, y_test_weather = train_test_split(x, y_weather, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train_temp)
```

```python
model_preds = model.predict(x_test)
print("Linear Regression MSE:", mean_squared_error(y_test_temp, model_preds))
```

```python
model_2 = DecisionTreeRegressor()
model_2.fit(x_train,y_train_temp)
```

```python
model_2_pred = model_2.predict(x_test)
print("Decision Tree Regressor MSE:", mean_squared_error(y_test_temp, model_2_pred))
```

```python
with open("linear_regression_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
with open("decision_tree_regressor.pkl", "wb") as f:
    pickle.dump(model_2, f)
```
